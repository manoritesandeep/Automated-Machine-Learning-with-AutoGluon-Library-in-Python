```python
from autogluon.multimodal import MultiModalPredictor
```

```python
# Download dataset

from autogluon.multimodal.utils.misc import shopee_dataset
```

```python
download_directory = "/content/sample_data"
```

```python
train_data, test_data = shopee_dataset(download_directory)
```

```python
train_data
```

```python
from IPython.display import Image, display
```

```python
display(Image(train_data["image"][0]))
```

```python
model_path = "image_classifier"
predictor = MultiModalPredictor(label="label",
                                path=model_path)
predictor.fit(train_data=train_data,
              time_limit=180)
```

```python
my_scores = predictor.evaluate(test_data, metrics=['accuracy'])
my_scores
```

```python
image_path = test_data.iloc[10]['image']

display(Image(image_path))
```

```python
predictions = predictor.predict({"image":[image_path]})
predictions
```

```python
prediction_proba = predictor.predict_proba({"image":[image_path]})
prediction_proba
```
